### Objective
Perform data cleaning and exploratory data analysis (EDA) on a Titanic-style dataset to explore relationships between variables and identify patterns.


Step 1 : Loading Data

```python
# Load Data
import pandas as pd

df = pd.read_csv("mock_titanic_dataset.csv")
df.head()
```

Step 2 : Cleaning Data

```python
# Check missing values
df.isnull().sum()

# Fill missing values
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Cabin'] = df['Cabin'].fillna('Unknown')
```

Step 3 : Exploratory Data Analysis

```python
# Survival Count
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Survived', hue='Survived', palette='Set2', legend=False)
plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
# Survival by Gender
sns.countplot(data=df, x='Sex', hue='Survived', palette='coolwarm')
plt.title("Survival by Gender")
plt.show()
```

```python
# Survival by Class
sns.countplot(data=df, x='Pclass', hue='Survived', palette='Set1')
plt.title("Survival by Passenger Class")
plt.show()
```

```python
# Age Distribution by Survival
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', bins=20)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
# Correlation Matrix
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title("Correlation Heatmap")
plt.show()
```
